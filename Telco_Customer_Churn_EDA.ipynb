```python
# Importing the seaborn library, which is built on top of matplotlib, for advanced and aesthetically pleasing data visualizations
import seaborn as sns
```

```python
# Loading the 'Customer_Churn.csv' dataset into a pandas DataFrame and displaying its contents
df = pd.read_csv('Customer_Churn.csv')
df  # Displays the first few rows of the dataset for inspection
```

```python
# Displaying the information about the DataFrame 'df' 
# This includes the number of entries, column data types, and non-null counts
df.info()
```

Replacing blanks with 0 as Tenure is 0 and no Total Charges are recorded

```python
# Replacing any empty spaces in the 'TotalCharges' column with 0 
df['TotalCharges'] = df['TotalCharges'].replace(' ', 0)

# Converting the 'TotalCharges' column from an object (string) to float for numerical analysis
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
# Displaying the information about the DataFrame, including column names, non-null values, and data types
df.info()
```

```python
# Check for missing values in the DataFrame by summing the number of null entries across all columns.
missing_values = df.isnull().sum().sum()
```

```python
# Generate descriptive statistics of the DataFrame, including count, mean, standard deviation, minimum, 
# quartiles, and maximum for each numerical column.
summary_statistics = df.describe()
```

```python
# Count the number of duplicate rows in the DataFrame to identify potential data quality issues.
duplicate_count = df.duplicated().sum()
```

```python
duplicate_count
```

```python
# Count the number of duplicate entries in the 'customerID' column to check for unique customer identification.
duplicate_customerID_count = df['customerID'].duplicated().sum()
```

```python
# Define a function to convert numeric values in the 'SeniorCitizen' column to categorical labels: 
# 1 is converted to 'Yes' and 0 is converted to 'No'.
def conv(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'

# Apply the conversion function to the 'SeniorCitizen' column of the DataFrame.
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

Converted 0 and 1 values of SeniorCitizen to Yes/No to make it easier to understand and analyze.

```python
# Display the first five rows of the DataFrame to get an overview of the data structure and content.
first_five_rows = df.head()
first_five_rows
```

```python
# Create a count plot to visualize the distribution of customers based on the 'Churn' status.
ax = sns.countplot(x='Churn', data=df)

# Add data labels to the bars for better readability.
ax.bar_label(ax.containers[0])

# Set the title for the plot.
plt.title('Count of Customers by Churn')

# Display the plot.
plt.show()
```

```python
# Set the figure size for the pie chart.
plt.figure(figsize=(3, 3))

# Group the DataFrame by 'Churn' status and count the number of customers in each category.
gb = df.groupby('Churn').agg({'Churn': 'count'})

# Create a pie chart to visualize the percentage of churned vs. non-churned customers.
plt.pie(gb['Churn'], labels=gb.index, autopct='%1.2f%%')

# Set the title for the pie chart.
plt.title('Percentage of Churned Customers')

# Display the pie chart.
plt.show()
```

From the given pie chart, we can conclude that 26.54% of customers are churned out.
Now, let's explore the reason behind it.

```python
# Set the figure size for the count plot.
plt.figure(figsize=(6, 4))

# Create a count plot to visualize the distribution of churned customers by gender.
cg = sns.countplot(x='gender', data=df, hue='Churn')

# Add data labels to the bars for better readability.
cg.bar_label(cg.containers[0])

# Set the title for the plot.
plt.title('Churn by Gender')

# Display the count plot.
plt.show()
```

```python
# Set the figure size for the count plot.
plt.figure(figsize=(6, 4))

# Create a count plot to visualize the distribution of churned customers based on senior citizen status.
sns.countplot(x='SeniorCitizen', data=df, hue='Churn')

# Set the title for the plot.
plt.title('Churn by Senior Citizen')

# Display the count plot.
plt.show()
```

```python
# Create a DataFrame to hold the percentage values for 'Churn' and 'SeniorCitizen'
df_percentage = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

# Plot a stacked bar chart to visualize the churn percentages by senior citizen status.
df_percentage.plot(kind='bar', stacked=True, figsize=(6, 4))

# Add percentage labels on the bars for better readability.
for n, x in enumerate([*df_percentage.index]):
    for (i, percentage) in enumerate(df_percentage.loc[x]):
        plt.text(x=n, y=df_percentage.iloc[n, :i+1].sum() - percentage/2, 
                 s=f'{percentage*100:.1f}%', ha='center', va='center', color='white', fontsize=12)

# Set the title and axis labels for the plot.
plt.title('Churn by Senior Citizen (as % of Total)', fontsize=14)
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')

# Display the stacked bar chart.
plt.show()
```

```python
# Set the figure size for the histogram plot.
plt.figure(figsize=(9, 4))

# Create a histogram to visualize the distribution of 'tenure' with respect to 'Churn' status.
sns.histplot(x="tenure", data=df, bins=72, hue="Churn")

# Display the histogram.
plt.show()
```

People who have used our services for a long time have stayed and people who have used our services 1 or 2 months have churned.

```python
# Set the figure size for the count plot.
plt.figure(figsize=(4, 4))

# Create a count plot to visualize the distribution of customers based on 'Contract' type, colored by 'Churn' status.
ax = sns.countplot(x="Contract", data=df, hue="Churn")

# Add data labels to the bars for better readability.
ax.bar_label(ax.containers[0])

# Set the title for the plot.
plt.title("Count of Customers by Contract")

# Display the count plot.
plt.show()
```

People who have month to month contract are likely to churn then from those who have 1 or 2 years or contract.

```python
# Retrieve the column names of the DataFrame as an array.
column_names = df.columns.values
```

```python
# Define a list of columns to visualize with count plots.
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this).
n_cols = 3
# Calculate the number of rows needed based on the number of columns.
n_rows = (len(columns) + n_cols - 1) // n_cols  

# Create subplots with the specified number of rows and columns.
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays).
axes = axes.flatten()

# Iterate over the specified columns and create count plots for each.
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue=df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')  # Set the title for each subplot.
    axes[i].set_xlabel(col)  # Set the x-axis label for each subplot.
    axes[i].set_ylabel('Count')  # Set the y-axis label for each subplot.

# Remove empty subplots (if any) to ensure a clean layout.
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust the layout to prevent overlapping of subplots.
plt.tight_layout()

# Display the plots.
plt.show()
```
